```python
from matplotlib.image import imread
from matplotlib.pyplot import subplots
```

```python
image = imread('figs/rose.jpg')

fig, ax = subplots()

ax.imshow(image);
```

```python
image.shape
```

```python
image_t = image.transpose((1, 0, 2))

fig, ax = subplots()

ax.imshow(image_t);
```

```python
fig, ax = subplots()

ax.imshow(image_t[:, :, 0], cmap='gray');
```

```python
fig, ax = subplots()

ax.hist(image_t[:, :, 0].ravel(), bins=50);
```

```python
threshold = 90

mask = image_t[:, :, 0] < threshold

image_masked = image_t[:, :, 0] * mask

fig, ax = subplots(ncols=3)

ax[0].imshow(image_t[:, :, 0], cmap='gray');

ax[1].imshow(mask, cmap='gray');

ax[2].imshow(image_masked, cmap='gray');

fig.tight_layout()
```

```python
image_new = image_t.copy()

gray_value = 130

image_new[mask, :] = gray_value

fig, ax = subplots()

ax.imshow(image_new);
```
